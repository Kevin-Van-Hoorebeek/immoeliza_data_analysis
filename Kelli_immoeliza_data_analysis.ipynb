```python
import pandas as pd
from IPython.display import display
# import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv('Data/Raw/property_information_Kelli.csv')
# Display the first few rows in more readable format
display(df.head(10))
df.shape
```

```python
df.columns.tolist()
```

### Remove 'apartment-block' from Property subtype

```python
df = df[df["Property subtype"].str.contains("apartment-block") == False]
display(df.head(10))
```

### Remove unwanted columns

```python
df.drop(labels=['Property subtype', 'Type of sale'], axis=1, inplace=True)
```

```python
display(df.head(10))
df.shape
```

### Remove unwanted listings with 'Number of bedrooms' as 'Nan'

```python
df.dropna(subset=['Number of bedrooms'], inplace=True)
display(df.head(40))
df.shape
```

```python
# Checking the number of Nan for living room....out of curiosity
nan_rows_livingarea = df[df['Living area m²'].isna()]
print(nan_rows_livingarea)
```

```python
# Filter to get only duplicate rows
duplicates = df[df.duplicated()]
print(duplicates)
```

### Create new csv file

```python
df.to_csv('Data/Raw/sanitized_property_listings_Kelli.csv', index=False)
```
