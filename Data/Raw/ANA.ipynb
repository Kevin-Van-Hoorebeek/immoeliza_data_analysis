MY QUESTION
What are the least/most expensive municipalities in Belgium/Wallonia/Flanders? (in terms of price per m², average price, and median price)

```python
import pandas as pd

# Download the cleaned data
file_path = r'C:\Users\Anastasiia\Documents\GitHub\immoeliza_data_analysis\Data\Cleaned\sanitized_property_listings_Kelli_final_sorted.csv'
df_cleaned = pd.read_csv(file_path)

# Let's review the first lines to understand the data structure
print(df_cleaned.head())

# Check whether there is a column with municipalities or cities
print(df_cleaned.columns)
```

We have to made column with the price per m²,.

```python
# Check whether there is a living area and a price
print(df_cleaned[['Price', 'Living area m²']].isnull().sum())

# Add a column with the price per m²
df_cleaned['Price per m²'] = df_cleaned['Price'] / df_cleaned['Living area m²']

# Check whether a new column has been added
print(df_cleaned[['Price', 'Living area m²', 'Price per m²']].head())
print(df_sorted.shape)
```

Need to group data by municipalities
Now that we have a column with the price per m², I can move on to grouping the data by municipality. This will allow me  to analyze the average, median and price per m² for each municipality.

```python
# Group data by locality data and calculate the average and median price
municipality_stats = df_cleaned.groupby('Locality data').agg(
 avg_price=('Price', 'mean'),
 median_price=('Price', 'median'),
 avg_price_per_m2=('Price per m²', 'mean')
).reset_index()

# Check the result
print(municipality_stats.head())
```

```python
# The most expensive municipalities by average price
most_expensive_avg = municipality_stats.sort_values(by='avg_price', ascending=False).head(10)

# The cheapest municipalities at an average price
cheapest_avg = municipality_stats.sort_values(by='avg_price').head(10)

print("Most expensive municipalities by average price:")
print(most_expensive_avg)

print("Cheapest municipalities by average price:")
print(cheapest_avg)
```

Our dataset only contains postcodes, but no explicit information about regions (Wallonia, Flanders, Brussels), using postcodes I can define regions. In Belgium, zip codes can be divided into regions, so I'am going to add a new "Region" column based on the zip code.

I have to create a dictionary that associates postal codes with regions. 
Postal codes in Belgium:

Brussels: 1000–1299

Flanders: 1500–1999 3000 – 3499 Flemish Brabant, 2000- 2999 Antwerp, 3500-3999 Limburg, 8000-8999 West Flanders , 9000-9999 East Flanders

Wallonia: 1300 -1499 Walloon Brabant, 7000-7999 Hainaut, 4000 -4999 Liège, 6600-6999 Luxembourg, 5000-5999 Namur

```python
import pandas as pd

# Function to determine the region by zip code
def get_region(postal_code):
    if 1000 <= postal_code <= 1299:
        return 'Brussels'
    elif 1300 <= postal_code <= 1499 or 4000 <= postal_code <= 7999:
        return 'Wallonia'
    elif 1500 <= postal_code <= 3999 or 8000 <= postal_code <= 8999 :
        return 'Flanders'
    else:
        return 'Unknown'

# Add a new 'Region' column based on zip code
df_cleaned['Region'] = df_cleaned['Locality data'].apply(get_region)

# We check the result
df_cleaned[['Locality data', 'Region']].head()
```

```python
# Filter the rows where the 'Region' column has the value 'Unknown'
unknown_regions = df_cleaned[df_cleaned['Region'] == 'Unknown']

# We check the result
print(unknown_regions[['Locality data', 'Region']])
print(df_cleaned[['Locality data', 'Region']])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# A function to determine the province by postal code
def get_province(postal_code):
    # Flanders
    if 1500 <= postal_code <= 1999 or 3000 <=postal_code <=3499:
        return 'Flemish Brabant'
    elif 2000 <= postal_code <= 2999:
        return 'Antwerp'
    elif 3500 <= postal_code <= 3999:
        return 'Limburg'
    elif 8000 <= postal_code <= 8999:
        return 'West Flanders'
    elif 9000 <= postal_code <= 9999:
        return 'East Flanders'
    
    # Wallonia
    elif 1300 <= postal_code <= 1499:
        return 'Walloon Brabant'
    elif 4000 <= postal_code <= 4999:
        return 'Liège'
    elif 5000 <= postal_code <= 5999:
        return 'Namur'
    elif 6000 <= postal_code <= 6599 or 7000 <= postal_code <= 7999:
        return 'Hainaut'
    elif 6600 <= postal_code <= 6999:
        return 'Luxembourg'
    
    # Brussels
    elif 1000 <= postal_code <= 1299:
        return 'Brussels'
    
    else:
        return 'Unknown'

# Adding a new column 'Province' based on the postal code
df_cleaned['Province'] = df_cleaned['Locality data'].apply(get_province)

# Save updated file with column 'Province'
output_file_path = r'C:\Users\Anastasiia\Documents\GitHub\immoeliza_data_analysis\Data\Cleaned\cleaned_data_with_region_and_price_per_m2.csv'
df_cleaned.to_csv(output_file_path, index=False)

print(f'File successfully saved as: {output_file_path}')
# Check the result
df_cleaned[['Locality data', 'Region', 'Province']].head()
print(df_cleaned.shape)



# We create a scatterplot to display the relationship between area and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Living area m²', y='Price', hue='Region', data=df_cleaned)
plt.title('Dependence of the price on the real estate area in different regions')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Create a barplot to compare the average price per square meter in each province
plt.figure(figsize=(12, 6))
sns.barplot(x='Province', y='Price per m²', data=df_cleaned, errorbar=None)
plt.title('Average price per square meter in the provinces')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pivot the data
price_pivot = df_cleaned.pivot_table(values='Price per m²', index='Locality data', columns='Province', aggfunc='mean')

plt.figure(figsize=(14, 8))
sns.heatmap(price_pivot, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'label': 'Average Price per m²'})
plt.title('Average Price per Square Meter by Municipality and Province')
plt.ylabel('Locality data')
plt.xlabel('Province')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Province', y='Price per m²', data=df_cleaned, inner='quartile')
plt.title('Price per Square Meter Distribution by Province')
plt.xticks(rotation=45)
plt.ylabel('Price per m²')
plt.xlabel('Province')
plt.show()
```

```python
# Calculate average and median prices per province
price_stats = df_cleaned.groupby('Province').agg({'Price per m²': ['mean', 'median']}).reset_index()
price_stats.columns = ['Province', 'Average Price per m²', 'Median Price per m²']

plt.figure(figsize=(12, 6))
sns.barplot(x='Province', y='Average Price per m²', data=price_stats, color='lightblue', label='Average Price')
sns.lineplot(x='Province', y='Median Price per m²', data=price_stats, color='orange', marker='o', label='Median Price')
plt.title('Average vs Median Price per Square Meter by Province')
plt.xticks(rotation=45)
plt.ylabel('Price per m²')
plt.xlabel('Province')
plt.legend()
plt.show()
```
