```python

import pandas as pd


# We load the table
file_path = r'C:\Users\Anastasiia\Documents\GitHub\immoeliza_data_analysis\Data\Raw\sanitized_property_listings_Kelli.csv'
df = pd.read_csv(file_path)

# Delete the columns 'Terrace surface m²' and 'Garden area m²'
df.drop(columns=['Terrace surface m²', 'Garden area m²'], inplace=True)

# Fill in the blanks in other columns
df['Living area m²'] = df['Living area m²'].fillna(df['Living area m²'].mean())
df['Number of facades'] = df['Number of facades'].fillna(df['Number of facades'].mode()[0])
df['Building condition'] = df['Building condition'].fillna('Not specified')

# Check for duplicates and remove them
df_cleaned = df.drop_duplicates()

# Save the cleaned table in a new CSV file in an accessible directory
output_file = r'C:\Users\Anastasiia\Documents\GitHub\immoeliza_data_analysis\Data\Raw\sanitized_property_listings_ANASTASIIA.csv'
df_cleaned.to_csv(output_file, index=False)

print(f"Cleared table saved to file: {output_file}")
```
